## Data Cleaning
### Map BORO ID to Names
- pd.Series.map(): Map to corresponding values

```python
boro_map = {1: 'Manhattan',
            2: 'The Bronx',
            3: 'Brooklyn',
            4: 'Queens',
            5: 'Staten Island'}

insp['BORO'] = insp['BORO'].map(boro_map)
insp.head()
```

### Map Cuisine codes
- pd.Series.to_dict(): Convert Series to {label->value} dict

```python
cuisine = pd.read_csv(path+'Cuisine.csv')
print cuisine.head()
cuisine.set_index('CUISINECODE', inplace=True) #Set CUISINECODE as index so that it's used as dict keys
print cuisine.head()
cuisine_dict = cuisine['CODEDESC'].to_dict()
print cuisine_dict
```

```python
insp['CUISINE'] = insp['CUISINECODE'].map(cuisine_dict)
insp.drop_duplicates(['CAMIS']).CUISINE.value_counts() #Count of restaurants per cuisine 
```

### How many restaurants by Boro and Cuisine?
- pd.pivot_table(): Create a spreadsheet-style pivot table

```python
insp_restaurants = insp.drop_duplicates(['CAMIS']) # Drop multiple inspections per restaurant
restaurant_count = pd.pivot_table(insp_restaurants, values='CAMIS', rows='CUISINE', cols='BORO', aggfunc=size)
restaurant_count.head()
```

```python
restaurant_count.fillna(0, inplace=True)
restaurant_count['All'] = restaurant_count.sum(axis=1)
restaurant_count
```

###Normalize by BORO to compare proportions
- pd.DataFrame.apply(): Applies a function to each row or column of the DataFrame

```python
restaurant_prop = restaurant_count.apply(lambda x: x * 1.0 / x.sum()) #Applied to each column by default
restaurant_prop.head()
```

```python
restaurant_prop.sort('All', ascending=False).head(15).plot(kind='barh', figsize=(12,10)) #Plot the top 15 cuisines by BORO
```

### How many restaurant inspections per zipcode in Manhattan?
####Groupby
- Aggregation: Compute a statistic for each group
- Transformation: Perform group-specific operations and return like-indexed data
- Filtration: Filter out groups based on an aggregation


```python
manhattan_insp = insp[insp['BORO'] == "Manhattan"][['ZIPCODE', 'CURRENTGRADE']] #Zipcodes and grades of Manhattan only
manhattan_insp.head()
```

####Aggregation

```python
manhattan_zipcodes = manhattan_insp.groupby('ZIPCODE') #Split into groups by zipcode
manhattan_zipcodes.CURRENTGRADE.size() #Aggregate to find a count of inspections by ZIPCODE
```

###Manhattan has some tiny zipcodes. Let's filter them out
####Filtration

```python
manhattan_insp = manhattan_insp.dropna(subset=['CURRENTGRADE']) #Drop rows missing a grade
print "Started with %s zipcodes" % manhattan_insp.ZIPCODE.nunique()
manhattan_insp = manhattan_insp.groupby('ZIPCODE').filter(lambda x: len(x) >= 500) #Keep Zipcodes with 500 or more inspections
print "Ended with %s zipcodes after filtering" % manhattan_insp.ZIPCODE.nunique()
manhattan_insp.sort('ZIPCODE').head() #Zipcode 10000 has been filtered out
```

### Which area of Manhattan has the best history of A-graded inspections?
#### pd.Series.agg(): Applies multiple aggregations at once
Aggregate the number of total inspections and number of A inspections per zip code

```python
manhattan_a_inspections = manhattan_insp.groupby('ZIPCODE').CURRENTGRADE.agg(
                                            {"A Inspections": lambda x: x.where(x=='A').count(),
                                               "Total Inspections": size})
manhattan_a_inspections.head()
```

```python
manhattan_a_inspections['% A Inspections'] = manhattan_a_inspections['A Inspections'] * \
                             1.0 / manhattan_a_inspections['Total Inspections']
manhattan_a_inspections.sort('% A Inspections')

#Bottom:
#Tribeca
#Chinatown
#Alphabet City

#Top
#Hell's Kitchen
#Chelsea
#Murray Hill
```

### How many inspections per day?
#### pd.DataFrame.resample(): Convert frequency of time-series data

```python
daily_inspections = insp.groupby('INSPDATE').CAMIS.count()
daily_inspections.index = pd.to_datetime(daily_inspections.index) #Convert strings to datetimes
print daily_inspections.head()
daily_inspections = daily_inspections.resample('D',how='sum')
daily_inspections.plot()
```

```python
monthly_inspections = daily_inspections.resample('MS',how='sum')
monthly_inspections = monthly_inspections.ix[monthly_inspections.index >= datetime.datetime(2009, 6, 1)]
monthly_inspections
```

```python
figure(figsize=(10,7))
monthly_inspections.plot()
```

### How often do inspections occur per restaurant?

```python
daily_insp = insp[['CAMIS','INSPDATE']]
daily_insp['DATE'] = pd.to_datetime(daily_insp['INSPDATE'])
daily_insp = daily_insp.sort(['CAMIS','DATE'])

inspection_gap = daily_insp.groupby('CAMIS').DATE.apply(lambda x: x.diff(1))
inspection_gap = inspection_gap.dropna()
inspection_gap = inspection_gap.map(lambda x: int(x / (1000000000 * 60 * 60 * 24)))
inspection_gap = inspection_gap.value_counts()
inspection_gap = inspection_gap.sort_index()

inspection_gap.plot(xlim=(0,500), figsize=(8,5))
```
